```python

import matplotlib.pyplot as plt
# la fonction ET logique
def ET(a,b):
    if a==1 and b==1 :
        return 1
    return 0
#la fonction OU logique
def OU(a,b):
    if a==0 and b==0:
        return 0
    return 1
#la fonction XOR logique
def XOR(a,b):
    if a==b:
        return 0
    return 1

def apprentissage(c):
    w1=0
    w2=0
    cmp=0
    b=0 # le biais
    
    alpha=0.7
    E=[] # on stocke les valeurs pour tracer la courbe EQM 
    
    x=[[0, 0], [0, 1], [1, 0], [1, 1]] # les entrées
    
    #on prend 20 comme le max d'iteration . on repete les iterations jusqu'a ce que la fonction appris
    for i in range(20):
        s=0
        
        # on calcule les valeurs d'entrer w1,w2 pour chaque elemenet du vecteur x
        for j in range(4):
            
            g=w1 * x[j][0] + w2 * x[j][1] + b # on calcul la somme pondéré des entrées de la fonction d’activation
            if(g>=0):
                y=1
            else:
                y=0
            
            #on calcule la valeur de d selon la fonction logique choisie par l'utilisateur
            if c==1:
                 d=ET(x[j][0],x[j][1])
            elif c==2:
                 d=OU(x[j][0],x[j][1])
            elif c==3:
                 d=XOR(x[j][0],x[j][1])
            else:
                break
            e=d-y
            # en applicant la regle de Windrow-Hoff sur w1,w2 et b
            w1=w1+alpha*e*x[j][0]
            w2=w2+alpha*e*x[j][1]
            b=b+alpha*e*1
            s=s+e*e # la somme de EQM

        E.append(s/2) # pour chaque iteration on ajoute une nouvelle valeur à E pour tracher la courbe EQM
        
 #pour chaque iteration on verifie la dernier valeur de E.si elle superieur à 0, on va incrementer le nombre d'iteration
        if E[-1]>0:
            cmp=cmp+1
    
    return w1,w2,b,cmp,E


while(1):
    print("-----------------FONCTION LOGIQUE A PRENDRE par le perceptron monocouche---------------")
    print("ET                       1")
    print("OU                       2")
    print("XOR                      3")
    print("SORTIR                   0")
    
    print("-------------------------------------------------------------------------------------------")
    
    inp=input("choisir la fonction logique ")
    inp1=int(inp)
    if inp1==1:
        print("apprentissage de la fonction logique **ET** ")
    elif inp1==2:
         print("apprentissage de la fonction logique **OU** ")
    elif inp1==3:
        print("apprentissage de la fonction logique **XOR** ")
    else:
        print("merci, BYE")
        break
    
    w1,w2,b,cmp,E=apprentissage(inp1)
    print("Entrainement termine ")
    print("** w1 = ",w1)
    print("** w2 = ",w2)
    print("** b= ",b)
    print("** la fonction a appris dans la",cmp,"eme iterations")
    plt.plot(E)
    plt.title("la courbe EQM en fonction du NI ")
    plt.show()
    print("LE TEST")
    choix2= ''
    while choix2 != 'n':
        inp=input("saisie le x1 : ")
        x1=int(inp)
        while x1!=1 and x1!=0: # on va repeter cette iteration jusqu'a l'utilisateur saisie une valeur de x1 binaire
            inp=input("x1 doit etre 0 ou 1.Resaisie le x1 ")
            x1=int(inp)
        inp=input("saisie le x2 ")
        x2=int(inp)
        while x2!=1 and x2!=0: # on va repeter cette iteration jusqu'a l'utilisateur saisie une valeur de x2 binaire
            inp=input("x2 doit etre 0 ou 1.Resaisie le x2 ")
            x2=int(inp)
        if(x1*w1+x2*w2+b<0):
            print("le resultat ====>0")
        else:

            print("le resultat ====>1")
            
        choix2=input("tester la fonction o/n")

        
    
    
        
```
